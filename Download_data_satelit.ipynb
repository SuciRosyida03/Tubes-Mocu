```python
def ahi_hsd_downloader(tstart,tend,dt,ch,res,sg,datdir):
  '''
  Input parameter:
    tstart,tend: waktu mulai dan akhir, string dengan format "yyyy,mm,dd,HH,MM"
    dt: interval waktu (dlm menit), integer
    ch: channel yg didownload, list string, eg: ['B01','B02','B03']
    res: resolusi channel yg didownload (menyesuaikan dengan ch), list string, eg: ['R10','R10','R05']
    sg: segmen data HSD yang didownload, list string, eg: ['04','05','06']
    datdir: direktori menyimpan data yg didownload, string
  '''

  import os
  import numpy as np
  from datetime import datetime, timedelta
  from getpass import getpass

  #data direktori
  if (not os.path.exists(datdir)):
    os.mkdir(datdir)
  #vektor waktu
  y1,m1,d1,h1,mn1 = map(int, tstart.split(','))
  y2,m2,d2,h2,mn2 = map(int, tend.split(','))
  t = np.arange(datetime(y1,m1,d1,h1,mn1), datetime(y2,m2,d2,h2,mn2), timedelta(minutes=dt)).astype(datetime)
  #entri USER dan PASSWORD JAXA
  #USER=input('USER:')
  #PASS = getpass('PASSWORD:')
  #loop download dengan wget
  for it in range(np.size(t)): #waktu
    for ic in range(np.size(ch)): #channel
      for iseg in range (np.size(sg)): #segmen
        fname='ftp://ftp.ptree.jaxa.jp/jma/hsd/'+t[it].strftime('%Y%m/%d/%H/HS_H08_%Y%m%d_%H%M_') + ch[ic] + '_FLDK_' + res[ic] + '_S' + sg[iseg]+ '10.DAT.bz2'
        !wget -c -w 3 --ftp-user=sucirosyida03_gmail.com --ftp-password=SP+wari8 $fname -P $datdir
```
